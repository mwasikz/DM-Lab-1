```python
#Convert the sunshine column’s zero value to the “Nan” type.
import numpy as np
df['sunshine'] = df['sunshine'].replace(0.0,float('Nan'))
df
```

```python
#Drop rows with the “Nan” value.
df=df[df["sunshine"].notnull()]
```

```python
df
```

```python
#Drop the column which you don’t need to measure similarities and distances and write
#why you drop that one.

df=df.drop(["rainfall"],axis=1)
df
```

Here I have dropped the "rainfall" column from the dataset. As the data from this column is either "yes/no" and not a numerical value it will difficult to measure the similarities and distances. 

```python
# Draw a scatter plot between two columns of which you think they have any correlation.
#And explain why.
import matplotlib.pyplot as plt
plt.scatter(df.cloud, df.sunshine)
plt.xlabel('cloud')
plt.ylabel('sunshine')
plt.show()
```

Here, I have drawn a scatter plot beteween the "cloud" and "sunshine" columns. From the dataset it can be seen that as the sunshine raises then there is less cloud and as the sunshine drops there is more cloud. Thats why there can be a correlation between these two columns. 

```python
#Drop other columns and print the dataset
df1=df.drop(["day","pressure","maxtemp","temparature","mintemp","dewpoint","humidity","winddirection","windspeed"],axis=1)
print(df1)
```

```python
#Calculate covariance and correlation among those two columns. 
covariance = np.cov(df1.cloud,df1.sunshine)
print(covariance[0,1])
```

```python
#Correlation
def pearson(x,y):
    correlation_matrix=np.corrcoef(x,y)
    return  correlation_matrix[0,1]
r=pearson(df1.cloud,df.sunshine)
print(r)
```

From the correlation value we can see that the value is very close to "-1". That means it is inversely correlated.
